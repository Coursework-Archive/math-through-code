#### Question 61

$$
x + (y - 1)^2 = 0
$$

$$
(y - 1)^2 = -x
$$

$$
y - 1 = \pm \sqrt{-x}
$$

$$
y = 1 \pm \sqrt{-x}
$$


Since the problem asks for the **bottom half** of the curve, we choose the negative branch:

$$
\boxed{y = 1 - \sqrt{-x}}
$$

For the square root to be defined:

$$
-x \ge 0 \;\Rightarrow\; x \le 0
$$

$$
\boxed{\text{Domain} = (-\infty,\,0]}
$$

#### Question 62

$$
x^2 + (y - 2)^2 = 4
$$

$$
(y - 2)^2 = 4 - x^2
$$

$$
y - 2 = \pm \sqrt{4 -x^2 }
$$

$$
y = 2 \pm \sqrt{4 - x^2 }
$$


Since the problem asks for the **top half** of the curve, we choose the positive branch:

$$
\boxed{y = 2 + \sqrt{4 - x^2 }}
$$

For the square root to be defined:

$$
4 - x^2 \ge 0 \;\Rightarrow\; -2 \le x \le 2
$$

$$
\boxed{\text{Domain} = [-2,\,2]}
$$
